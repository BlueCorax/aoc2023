```python
test_input = """...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#.....""".splitlines()
```

```python
GALAXY = '#'
def parse_galaxies(map_string, empty_space_mult = 2):
    empty_rows = [i for i in range(len(map_string)) if not GALAXY in map_string[i]]
    empty_cols = [i for i in range(len(map_string[0])) if not GALAXY in [r[i] for r in map_string]]
    galaxies = []
    y = 0
    for j in range(len(map_string)):
        x = 0
        if j in empty_rows: y += empty_space_mult - 1
        for i in range(len(map_string[j])):
            if i in empty_cols: x += empty_space_mult - 1
            if map_string[j][i] == GALAXY:
                galaxies.append((x, y))
            x += 1
        y += 1
    return galaxies
```

```python
parse_galaxies(test_input)
```

```python
def shortest_path(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])
```

```python
def sum_paths(inp, empty_space_mult = 2):
    galaxies = parse_galaxies(inp, empty_space_mult)
    print(galaxies)
    path_sum = 0
    for i in range(len(galaxies) - 1):
        for j in range(i + 1, len(galaxies)):
            path_sum += shortest_path(galaxies[i], galaxies[j])
    return path_sum
```

```python
def part1(inp):
    return sum_paths(inp)
```

```python
part1(test_input)
```
